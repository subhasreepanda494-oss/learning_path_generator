```python
def generate_learning_path(skill):
    roadmap = f"""
 LEARNING ROADMAP FOR {skill.upper()}

 1. BEGINNER LEVEL (0-1 Month)
- Understand basic concepts of {skill}
- Learn fundamental theory
- Watch free YouTube tutorials (Channels: FreeCodeCamp, Apna College)

 Resources:
- YouTube
- W3Schools / GeeksforGeeks

 2. INTERMEDIATE LEVEL (1-3 Months)
- Work on small projects
- Practice real-world problems
- Learn libraries/tools related to {skill}

 Resources:
- Kaggle (for Data Science)
- GitHub projects
- Online courses (Coursera, Udemy - free content)

 3. ADVANCED LEVEL (3-6 Months)
- Build real-world applications
- Contribute to open-source
- Create portfolio

Resources:
- Advanced tutorials
- Documentation

BONUS TIPS:
✔ Practice daily
✔ Build projects
✔ Stay consistent

Estimated Time: 3-6 Months
"""
    return roadmap


# User Input
skill = input("Enter a skill (e.g., Python, AI, Data Science): ")

# Output
result = generate_learning_path(skill)
print(result)

```
